```python
import pandas as pd
import io

# This creates a sample version of the oil_prices.csv file mentioned in your notes
csv_content = """date,year,month,day,Price,Vol
2021-12-01,2021,12,1,70.12,450K
2021-12-02,2021,12,2,69.50,420K
2021-12-03,2021,12,3,71.20,
2021-12-04,2021,12,4,72.10,380K
2021-01-01,2021,01,1,54.77,310K
2020-05-10,2020,05,10,29.38,200K
2019-11-15,2019,11,15,63.21,
2022-03-14,2022,03,14,117.25,600K
2021-12-25,2021,12,25,74.17,350K
2021-12-31,2021,12,31,75.10,
"""

# Save this to a file named oil_prices.csv
with open('oil_prices.csv', 'w') as f:
    f.write(csv_content)

print("File 'oil_prices.csv' has been created successfully!")
```

```python
# 1. Load the data
path = "oil_prices.csv"
df = pd.read_csv(path)

# 2. Convert columns to correct types (as per your notes)
df["Vol"] = pd.to_numeric(df["Vol"].str.replace('K', ''), errors="coerce")
df['day'] = df['day'].astype('category')
df['month'] = df['month'].astype('category')
df['year'] = df['year'].astype('category')

# 3. View the data
print("--- Top 5 Rows ---")
print(df.head(5))

# 4. Filter: December 2021
print("\n--- December 2021 Records ---")
dec_2021 = df.loc[(df.year == 2021) & (df.month == 12), :]
print(dec_2021)

# 5. Descriptive Statistics
print("\n--- Statistics for Numeric Columns ---")
print(df.describe())

print("\n--- Statistics for Categorical Columns ---")
print(df.describe(include=['category']))

# 6. Count missing values in Vol
missing_vol = df['Vol'].isnull().sum()
print(f"\nNumber of missing records in Vol: {missing_vol}")
```
